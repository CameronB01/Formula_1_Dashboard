```python
from matplotlib import pyplot as plt
import fastf1
import fastf1.plotting

fastf1.plotting.setup_mpl()

session = fastf1.get_session(2023, 'Qatar', 'Q')

session.load()
fast_leclerc = session.laps.pick_driver('LEC').pick_fastest()
lec_car_data = fast_leclerc.get_car_data()
t = lec_car_data['Time']
vCar = lec_car_data['Speed']

fast_verst = session.laps.pick_driver('VER').pick_fastest()
verst_car_data = fast_verst.get_car_data()
vt = verst_car_data['Time']
vvCar = verst_car_data['Speed']

fig, ax = plt.subplots()
ax.plot(t, vCar, label='Lec')
ax.plot(vt, vvCar, label='Ver')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.set_title('Leclerc is')
ax.legend()
plt.show()
```
